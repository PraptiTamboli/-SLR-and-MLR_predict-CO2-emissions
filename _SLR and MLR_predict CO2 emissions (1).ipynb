```python
# SLR Model(Simple Linear Regression)
```

```python
df.columns
```

To build a Simple Linear Regression (SLR) model, we must have an independent variable and a dependent variable. For our SLR model, we are going to take Engine size as the independent variable and undoubtedly CO2 emissions as the dependent variable.

```python
x_var = df[['ENGINESIZE']] # Independent variable
y_var = df['CO2EMISSIONS'] # Dependent variable
```

```python
# Statsmodels
```

```python
! pip install termcolor
```

```python
import statsmodels.api as sm
from termcolor import colored as cl 

slr_model = sm.OLS(y_var, x_var) # Ordinary Least Squares 
slr_reg = slr_model.fit()
```

Code Explanation: At first, we have imported our primary package which is “statsmodels.api”. Next, we have defined a variable “slr_model” to store our Ordinary Least Squares (OLS) model, and finally, we stored our fitted model to a variable “slr_reg”.

Now let’s see the results of our model’s performance.

```python
print(cl(slr_reg.summary(),attrs=['bold']))
```


When analyzing our results summary, we can notice that the R-squared of the model is 0.952 (95.2%) which clearly reveals that our model is doing well and can be used for real-world cases for solving problems.

```python
# Scikit-learn
```

Like how we used the OLS model in statsmodels, using scikit-learn, we are going to use the ‘train_test_split’ algorithm to process our model. 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_var,y_var,test_size=0.3,random_state=0)

lr = LinearRegression()
lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)
```

Code Explanation: Firstly, we are importing our primary packages which are “LinearRegression” and “train_test_split”. Using the train_test_split algorithm, we are classifying the training dataset and the testing dataset whose size is 30% of the original dataset. Inside the train_test_split algorithm, I’ve passed a command “random_state = 0” which means, there should be no automatic random shuffling of data when classifying train and test data. Next, we are storing our linear model to the variable “lr” and fitting the model to the variables. Finally, we are storing the predicted values to the variable “y_pred”.

Now to check the accuracy of our scikit-learn model, we are going to calculate the slope and intercept and fit that values into our model also, we are going to calculate the R-Squared value of the model. 

```python
slr_slope = lr.coef_
slr_intercept = lr.intercept_

sns.scatterplot(x = 'ENGINESIZE', y = 'CO2EMISSIONS', data = df, s = 150, alpha = 0.3, edgecolor = 'white')
plt.plot(df['ENGINESIZE'], slr_slope*df['ENGINESIZE'] + slr_intercept, color = 'r', linewidth = 3)
plt.title('Engine size vs C02 Emissions', fontsize = 16)
plt.ylabel('CO2 Emissions', fontsize = 12)
plt.xlabel('Engine Size', fontsize = 12)

#plt.savefig('enginesize_co2_fit.png')
```

Now let’s calculate the R-squared value of our model by scikit-learn.

```python
print(cl('R-Squared :',attrs=['bold']),lr.score(x_test,y_test))
```

We can notice that the value of R-squared in the scikit-learn model is different from the statsmodels model. This is because we didn't add a constant value to the independent variable in the statsmodels model. In the upcoming MLR model, we will be adding a constant value to the independent variable in the statsmodels.

```python
# MLR Model
```

We have successfully created our SLR model using both statsmodel package and the scikit-learn package. Now let’s dive into building the Multiple Linear Regression (MLR) model.

To build a Multiple Linear Regression (MLR) model, we must have more than one independent variable and a dependent variable. For our MLR model, we are going to take four independent variables and undoubtedly CO2 emissions as the dependent variable.

```python
df.columns
```

```python
x1_var = df[['ENGINESIZE','FUELCONSUMPTION_COMB','FUELCONSUMPTION_HWY','FUELCONSUMPTION_CITY']]
y_var = df['CO2EMISSIONS'] # Dependent Variable
```

Remember that, adding more and more independent variables to the model might result in “Overfitting”. In our CO2 data, we have only a small number of attributes but in case of huge data, we must be more cautious about picking independent variables. So, it is highly recommended to choose only relevant independent variables to the dependent variable.

```python
# Statsmodels
```

```python
sm_X1_var = sm.add_constant(x1_var)

mlr_model = sm.OLS(y_var,sm_X1_var)
mlr_reg = mlr_model.fit()
```

You can notice that we have added a constant value to our independent variable.

Now that we have fitted our model and let’s view the results summary.

```python
print(cl(mlr_reg.summary(), attrs=['bold']))
```

When analyzing our results summary, we can notice that the R-squared of the model is 0.859 (85.9%) and this value is derived by including the constant value of the independent variable. So we can say that this model can be used to solve real-world cases.

```python
# Scikit-learn
```

The code implementation and the algorithm used are the same as the SLR model but adding extra attributes to the independent variable.

```python
x_train,x_test,y_train,y_test = train_test_split(x1_var,y_var,test_size=0.3,random_state=0)

lr = LinearRegression()
lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)
```

To check the accuracy of the scikit-learn model, we can calculate the R-squared score and we can introduce a new way which is by distribution plot. 

```python
print(cl('R-Squared :', attrs=['bold']), lr.score(x_test,y_test))
```

We can observe that the R-squared value of the scikit-learn model is almost similar to the statsmodels model whose value is 0.84. This is because we have added a constant value to the independent variable while building the statsmodels MLR model.

The second method to check the accuracy of the MLR scikit-learn model is by constructing a distribution plot by combining the predicted values and the actual values.

```python
sns.distplot(y_pred, hist = False, color = 'r', label = 'Predicted Values')
sns.distplot(y_test, hist = False, color = 'b', label = 'Actual Values')
plt.title('Actual vs Predicted Values', fontsize = 16)
plt.xlabel('Values', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)
plt.legend(loc = 'upper left', fontsize = 13)

# plt.savefig('ap.png'))
```

This distribution plot reveals that our prediction values have performed almost precisely to our actual values but there are some outliers that can be noticed. This is because we have built a very basic model on Linear Regression to precisely predict the outcomes.
